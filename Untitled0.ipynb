```python
import numpy as np

arr = np.linspace(0, 9, 10, dtype=int)
print("Original Array:", arr)
```

```python

arr_modified = arr.copy()
arr_modified[arr_modified % 2 != 0] = -1
print("Modified Array:", arr_modified)
print("Original Array (unchanged):", arr)
```

```python

arr_2d = arr.reshape(2, -1)
print("2D Array:\n", arr_2d)
```

```python

even_sum = 0
for num in arr:
    if num % 2 == 0:
        even_sum += num

print("Sum of all even numbers:", even_sum)
```
